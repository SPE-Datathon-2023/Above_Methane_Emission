```python
# WARNING : this depends on the resolution of your screen !
# Dpi is set to 151 divided by aspect ratio for some reason...

def MergeandSavePNG(r_path, w_path, rgb, ctr):
    img_r = mpimg.imread(r_path + rgb)
    img_c = mpimg.imread(r_path + ctr)
    plt.axis('off')
    plt.imshow(img_r)
    plt.imshow(img_c)
    f_name = rgb[:-7] + '.png'
    plt.savefig(w_path + f_name, bbox_inches='tight', pad_inches=0,  dpi=151/2.84)
    plt.clf()
    #print(f_name)
    #plt.show
    return f_name
    
```

```python
#plume_list['new_png_file_name'] = ""
R_path = 'Purple_Plumes/permian_2019/permian_2019_rgb_ctr/'
W_path = 'Purple_Plumes/generated_dataset/'

for i in range(2000, len(plume_list)):
    RGB = plume_list['new_rgb_file_name'].loc[i]
    CTR = plume_list['new_ctr_file_name'].loc[i]    

    plume_list['new_png_file_name'].loc[i] = MergeandSavePNG(R_path, W_path, RGB, CTR)
    if i%100 == 0:
        print(i)
```
