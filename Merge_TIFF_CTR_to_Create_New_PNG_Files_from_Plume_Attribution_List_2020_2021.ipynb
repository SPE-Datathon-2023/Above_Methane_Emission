```python
# WARNING : this depends on the resolution of your screen !
# Dpi is set to 151 divided by aspect ratio for some reason...

def MergeandSavePNG(r_path, w_path, rgb, ctr):
    img_r = mpimg.imread(r_path + rgb)
    #img_r = img_r.resize(size)
    img_c = mpimg.imread(r_path + ctr)
    #img_c = img_c.resize(size)
    plt.axis('off')
    plt.imshow(img_r)
    plt.imshow(img_c)
    f_name = rgb[:-8] + '.png'
    plt.savefig(w_path + f_name, bbox_inches='tight', pad_inches=0,  dpi=151/2.84)
    plt.clf()
    #print(f_name)
    #plt.show
    return f_name
    
```

```python
plume_list['new_png_file_names'] = ""
R_path = 'Purple_Plumes/CM_2020_2021_Images/carbonmapper_ch4_rgb_geotiffs_2020_2021/'
W_path = 'Purple_Plumes/CM_2020_2021_Images/generated_images/'

for i in range(2087, len(plume_list)):
    RGB = plume_list['rgb_file_names'].loc[i]
    CTR = plume_list['ctr_file_names'].loc[i]    

    plume_list['new_png_file_names'].loc[i] = MergeandSavePNG(R_path, W_path, RGB, CTR)
    if i%200 == 0:
        print(i)
```
