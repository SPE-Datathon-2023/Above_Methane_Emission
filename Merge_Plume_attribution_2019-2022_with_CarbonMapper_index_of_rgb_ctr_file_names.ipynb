```python
def Split_rgb_ctr_names(entry):
    part = entry.partition(', ')
    Ls = part[0][2:-1]
    Rs = part[2][1:-2]
    if 'rgb' in Ls:
        RGB = Ls
        CTR = Rs
    else:
        RGB = Rs
        CTR = Ls
    return RGB, CTR
```

```python
names = file_names_att['file_names']
rgb_files = []
ctr_files = []
for name in names:
    rgb, ctr = Split_rgb_ctr_names(name)
    rgb_files.append(rgb)
    ctr_files.append(ctr)
    
file_names_att['rgb_file_names'] = rgb_files
file_names_att['ctr_file_names'] = ctr_files
```
