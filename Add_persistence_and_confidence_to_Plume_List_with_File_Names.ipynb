```python
import pandas as pd
```

```python
plume_att = pd.read_csv('permian_source_list_2019 Jeremy Zhao.csv')
plume_att.drop(['Unnamed: 0'], axis = 1, inplace =True)
plume_att.head()
```

```python
plume_att.tail()
```

```python
plume_list = pd.read_csv('permian_plume_list_2019_with_facility_sector_oldpng_and_new_ctr_tif_file_names.csv')
plume_list.drop(['Unnamed: 0'], axis = 1, inplace =True)
plume_list.head()
```

```python
plume_list.shape, plume_att.shape
```

```python
id_by_key = {}
s_pers_conf = []

for k in range(len(plume_att)):
    source = plume_att['source_id'].loc[k]
    if source in id_by_key:
        print(source)
    else:
        s_pers_conf = [ plume_att['source_persistence'].loc[k], plume_att['confidence_in_persistence'].loc[k] ]
        id_by_key[source] = s_pers_conf

#id_by_key      
```

```python
plume_list['source_persistence'] = ""
plume_list['confidence_in_persistence'] = ""

for j in range(len(plume_list)):
    source = plume_list['source_id'].loc[j]
    plume_list['source_persistence'].loc[j] = id_by_key[source][0]
    plume_list['confidence_in_persistence'].loc[j] = id_by_key[source][1]

plume_list.head()
```

```python
plume_list.to_csv('permian_plume_list_2019_facility_sector_oldpng_and_ctr_tif_names_pers_conf.csv', index = False)
```
