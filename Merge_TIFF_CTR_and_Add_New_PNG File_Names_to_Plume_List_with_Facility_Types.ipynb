```python
import pandas as pd
```

```python
Path = "Purple_Plumes/permian_2019\permian_2019/"
```

```python
file_names = pd.read_csv('permian_tiff_rgb_ctr_files.csv')
file_names.head()
```

```python
plume_list = pd.read_csv('permian_plume_list_2019_with_facility_sector_and_file_names.csv')
plume_list.head()
```

```python
plume_list.shape, file_names.shape
```

```python
plume_list['new_rgb_file_name'] = ""
plume_list['new_ctr_file_name'] = ""

rgb_file_list = []
ctr_file_list = []

for candidate in plume_list['candidate_id']:
    rgb, ctr =0, 0 #set file found flag to zero - reduce number of iterations
    l = len(candidate)
    for name in file_names['permian_tiff_rgb_ctr_files']:
        if ctr * rgb == 0:
            
            if candidate == name[0:l]:
                if rgb == 0:
                    if 'rgb' in name:
                        rgb_file_list.append(name)
                        rgb = 1
                else:
                    pass

                if ctr == 0:
                    if 'ctr' in name:
                        ctr_file_list.append(name)
                        ctr = 1
                    else:
                        pass
                else:
                    pass
            else:
                pass
        else:
            break
        
    # Check if rgb and ctr file name found, if not enter 'none'
    if rgb == 0:
        rgb_file_list.append('none')
    
    if ctr == 0:
        ctr_file_list.append('none')
               
            
plume_list['new_rgb_file_name'] = rgb_file_list
plume_list['new_ctr_file_name'] = ctr_file_list

plume_list.head()
```
